```python
from math import *
from sympy import *
from sympy.plotting import plot
init_printing()
```

```python
x = Symbol('x')
```

```python
y = 5*x**2 + 10*x + 150
```

```python
# Построить график
plot(y)
print(f'f(x) = {y}')
```

```python
plot(diff(y))
diff(y)
```

```python
# Определить корни
solve(y)
x1, x2 = solve(y)
```

```python
print(f' x1 = {round(x1.evalf(), 2)}')
```

```python
print(f' x2 = {round(x2.evalf(), 2)}')
```

```python
# Вычислить вершину
solve(diff(y))
```

```python
# Найти интервалы, на которых функция возрастает
solve(y > 0)
```

```python
# Найти интервалы, на которых функция убывает
solve(y < 0)
```

```python
# Определить промежутки, на котором f > 0
solve(diff(y) > 0)
```

```python
# Определить промежутки, на котором f < 0
solve(diff(y) < 0)
```
